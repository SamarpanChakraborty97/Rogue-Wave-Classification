```python
import os
import numpy as np
import pickle
```

```python
file_path = os.getcwd() + "/" + "Buoy_meta_data.npy"
file_contents = np.asarray(np.load(file_path, allow_pickle=True))
print(file_contents)
```

```python
array_length = file_contents.shape[0]
# print(type(file_contents[0]['name']))

for i in range(array_length):
    for j in file_contents[i].keys():
        if j=='name':
            print(f"Buoy number: {file_contents[i][j]}")
            
        elif j=='WaterDepth':
            print(f"Depth: {file_contents[i][j]}")
        elif j=='DeployLatitude':
            print(f"Latitude: {file_contents[i][j]}")
        elif j=='DeployLongitude':
            print(f"Longitude: {file_contents[i][j]}")

    print('\n')
```

```python
buoy_numbers_starts = []
for i in range(array_length):
    buoy_number = file_contents[i]['name'][:3]
    # print(buoy_number, end= ' ')
    if i==0:
        buoy_numbers_starts.append(i)
    elif buoy_number != file_contents[i-1]['name'][:3]:
        buoy_numbers_starts.append(i)

print(buoy_numbers_starts)
```

```python
len(buoy_numbers_starts)
```

```python
import statistics

names = []
latitudes = [ ]
longitudes = []
depths = []

for i in range(len(buoy_numbers_starts)):
    name_buoy = file_contents[buoy_numbers_starts[i]]['name'][:3]
    names.append(name_buoy)

    if i <= len(buoy_numbers_starts)-2:
        depth_buoy = []
        latitude_buoy = []
        longitude_buoy = []
        
        for j in range(buoy_numbers_starts[i], buoy_numbers_starts[i+1]):
            if file_contents[j]['DeployLatitude'] != 0.0:
                depth_buoy.append(file_contents[j]['WaterDepth'])
                latitude_buoy.append(file_contents[j]['DeployLatitude'])
                longitude_buoy.append(file_contents[j]['DeployLongitude'])
            
        depths.append(statistics.mean(depth_buoy))
        latitudes.append(statistics.mean(latitude_buoy))
        longitudes.append(statistics.mean(longitude_buoy))

    else:
        depth_buoy = []
        latitude_buoy = []
        longitude_buoy = []
        
        for j in range(buoy_numbers_starts[i], array_length):
            if file_contents[j]['DeployLatitude'] != 0.0:
                depth_buoy.append(file_contents[j]['WaterDepth'])
                latitude_buoy.append(file_contents[j]['DeployLatitude'])
                longitude_buoy.append(file_contents[j]['DeployLongitude'])

        depths.append(statistics.mean(depth_buoy))
        latitudes.append(statistics.mean(latitude_buoy))
        longitudes.append(statistics.mean(longitude_buoy))


buoy_info = np.asarray([names, depths, latitudes, longitudes]).T
np.savetxt("buoy_information.csv", buoy_info.astype(float), delimiter = ",")   
```
